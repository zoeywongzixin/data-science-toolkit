```python
import numpy as np
```

```python
# 1. Create a numpy array of size 2 x 2
a = np.array([[1,2],[3,4]])
print(a)
```

```python
# 2. Create a random numpy array of size 3 x 4
b = np.random.random((2,2))
print(b)
```

```python
# 3. Create a numpy array of ones and a numpy array of zeros
n1 = np.ones((3,3))
print(n1)

print("")

n2 = np.zeros((3,3))
print(n2)
```
